```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
```

```python
print("a. Most of the people who survived were from Pclass = 3:", survival_by_class.idxmax() == 3)
```

```python
print("b. Most of the people who survived were from Pclass = 1:", survival_by_class.idxmax() == 1)
```

```python
print("c. Most of the people who survived were from Pclass = 2:", survival_by_class.idxmax() == 2)
```

```python
pclass_3_ratio = df[df['Pclass'] == 3]['Survived'].value_counts(normalize=True).get(1, 0) / df[df['Pclass'] == 3]['Survived'].value_counts(normalize=True).get(0, 0)
print("d. The ratio of survived = 1 to survived = 0, for Pclass = 3 is greater than 1:", pclass_3_ratio > 1)
```
